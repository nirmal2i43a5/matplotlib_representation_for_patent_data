```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame with your provided data
data = {
    '2004': [22, 3, 2, 0, 1, 0, 0, 0, 0],
    '2005': [108, 11, 3, 1, 2, 1, 0, 1, 0],
    '2006': [166, 6, 2, 1, 0, 0, 0, 0, 0],
    '2007': [180, 19, 2, 1, 0, 0, 0, 0, 0],
    '2008': [270, 23, 2, 1, 1, 0, 0, 2, 0],
    '2009': [308, 21, 4, 4, 0, 3, 0, 0, 0],
    '2010': [368, 24, 0, 3, 0, 1, 1, 0, 0],
    '2011': [424, 25, 3, 4, 2, 2, 3, 0, 0],
    '2012': [471, 38, 3, 7, 1, 2, 0, 0, 0],
    '2013': [613, 37, 7, 5, 1, 2, 0, 0, 0],
    '2014': [644, 38, 3, 8, 2, 3, 0, 0, 0],
    '2015': [878, 55, 7, 13, 1, 0, 1, 0, 0],
    '2016': [1084, 93, 9, 16, 3, 0, 3, 0, 0],
    '2017': [1117, 105, 8, 32, 2, 1, 2, 0, 0],
    '2018': [1208, 134, 6, 42, 2, 2, 2, 0, 0],
    '2019': [1529, 192, 14, 52, 3, 2, 0, 0, 0],
    '2020': [1547, 199, 5, 66, 2, 3, 0, 0, 0],
    '2021': [1377, 185, 12, 76, 0, 2, 0, 0, 0],
    '2022': [1168, 95, 11, 39, 0, 1, 0, 0, 0]
}

df = pd.DataFrame(data, index=['US', 'EP', 'KOREA', 'china', 'canada', 'japan', 'israel', 'UK', 'france'])

# Create a figure and axis for each country
for country in df.index:
    fig, ax = plt.subplots(figsize=(8, 8))  # Adjust size as needed

    # Get the data for the selected country
    counts = df.loc[country]
    years = counts.index

    # Plot the pie chart with counts
    wedges, texts, autotexts = ax.pie(counts, labels=years, autopct='%1.0f%%', startangle=140, wedgeprops={'edgecolor': 'black'})

    # Set aspect of the pie chart to be equal
    ax.axis('equal')

    # Set title
    ax.set_title(f'Patent Counts for {country}')

    # Add counts to the pie chart
    for i, autotext in enumerate(autotexts):
        autotext.set_text(f'{counts[i]}')

    # Show the pie chart
    plt.show()
```
